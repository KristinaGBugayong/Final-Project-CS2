```python
import json


with open("/content/weather.json", "r") as file:
 weather_records = json.load(file)

while True:
  print("""============================
CLIMATE CHANGE CALCULATOR
============================
[1] Find the day with the highest wind speed
[2] Generate full climate risk summary
[3] Calculate average wind speed for the week
[4] Exit
  """)
  choice = int(input("Enter choice from the list:"))
  print("\n")
  if choice == 1:
    max_wind = 0
    windiest_day = ""
    wind_status = ""
    for record in weather_records:
      if record["wind_speed_kmh"] > max_wind:
        max_wind = record["wind_speed_kmh"]
        windiest_day = record["date"]
    print(f"Highest wind speed recorded: {max_wind}")
    print(f"Windiest day: {windiest_day}")
    if max_wind >= 118:
      print("Super typhoon-force winds!")
    elif max_wind >= 89:
      print("Typhoon-force winds - very dangerous!")
  elif choice == 2:
    print("\nCLimate Risk Summarry Report \n----------------------------")
    print(f"Days Analyzed: {len(weather_records)}")
    average_temp = 0
    total_rainfall = 0
    rainiest_day = ""
    highest_rainfall = 0
    max_wind = 0
    windiest_day = ""
    dominant_condition = ""
    conditions = []
    for record in weather_records:
      temp = record["temperature"]
      average_temp += temp["avg"]
      total_rainfall += record["rainfall_mm"]
      if record["rainfall_mm"] > highest_rainfall:
                  highest_rainfall = record["rainfall_mm"]
                  rainiest_day = record["date"]
      if record["wind_speed_kmh"] > max_wind:
        max_wind = record["wind_speed_kmh"]
        windiest_day = record["date"]
      conditions.append(record["condition"])
    average_temp = average_temp/len(weather_records)
    print(f"Average Temperature: {average_temp}")
    print(f"Total Rainfall: {total_rainfall}")
    print(f"Highest Rainfall: {highest_rainfall}")
    print(f"Wettest day: {rainiest_day}")
    print(f"Highest windspeed: {max_wind}")
    print(f"Windiest day: {windiest_day}")
    print(f"Most prominient condition: {max(set(conditions), key=conditions.count)}")

  elif choice == 3:
    total_wind = 0
    day_count = 0
    for record in weather_records:
      total_wind += record["wind_speed_kmh"]
      day_count += 1
    avg_wind_week = total_wind / day_count
    print(f"Average wind speed this week: {avg_wind_week} kph")
    if avg_wind_week > 40:
      print("Secure loose objects")
    if avg_wind_week < 15:
      print("Calm conditions - favorable for outdoor activities")

  elif choice == 4:
    print("""Thank you for using Climate Change Calculator.
    Stay informed. Stay safe. Protect our planet.""")
    break

  else:
    print("Invalid choice! Please enter a number from 1 to 7.")

  if choice != 4:
    print("\nPress enter to return to main menu.")
    input("")
```
